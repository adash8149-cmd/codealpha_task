```python
from deep_translator import GoogleTranslator
import ipywidgets as widgets
from IPython.display import display
```

```python
text_input = widgets.Textarea(
    placeholder="Enter text here...",
    description="Text:",
    layout=widgets.Layout(width='600px', height='120px')
)
```

```python
source_lang = widgets.Dropdown(
    options=[
        ('Auto Detect', 'auto'),
        ('English', 'en'),
        ('Hindi', 'hi'),
        ('French', 'fr'),
        ('German', 'de'),
        ('Spanish', 'es')
    ],
    value='auto',
    description='Source:'
)

target_lang = widgets.Dropdown(
    options=[
        ('English', 'en'),
        ('Hindi', 'hi'),
        ('French', 'fr'),
        ('German', 'de'),
        ('Spanish', 'es')
    ],
    value='hi',
    description='Target:'
)
```

```python
translate_button = widgets.Button(
    description="Translate",
    button_style='success'
)

clear_button = widgets.Button(
    description="Clear",
    button_style='warning'
)

output = widgets.Output()
```

```python
def translate_text(b):

    output.clear_output()

    text = text_input.value

    translated = GoogleTranslator(
        source=source_lang.value,
        target=target_lang.value
    ).translate(text)

    with output:
        print("Original Text:")
        print(text)
        print()
        print("Translated Text:")
        print(translated)

translate_button.on_click(translate_text)
```

```python
def clear_text(b):

    text_input.value = ""

    output.clear_output()

clear_button.on_click(clear_text)
```

```python
display(text_input)
display(source_lang)
display(target_lang)
display(translate_button)
display(clear_button)
display(output)
```
